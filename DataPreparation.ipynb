### This section converts the column headers to a list, then uses this as the keys in a dictionary. for the values, we compute the necessary values and append to the dictonary based on the corresponding key (the name of the genetic mutation). the values of M(also T), C,NC computed earlier are also appended

```python
genetic_mutations = df.drop(columns='class').columns.tolist()

result = {}
for i in genetic_mutations:
    PC = C[i]/C_total
    PNC = NC[i]/NC_total
    Sub_CNC = PC-PNC
    if not PNC == 0:
        Div_CNC = PC/PNC
    
    
    result[i] = [M[i], C[i], NC[i], PC, PNC, Sub_CNC, Div_CNC]

resultdf = pd.DataFrame(result)

resultdf.index = ['T', 'C', 'NC', '%C', '%NC', '%C-%NC', '%C/%NC']
feature_table = resultdf.transpose()
# for i in mutation_sum.items:
    
```

## Creating a Confusion Matrix

### find the actual values

```python
actual_values = df['class'].str.startswith('C').astype(int)
actual_values
```

### find the predicted values for the highest value in table 1

```python
PPP = table1.head(2).index[0]
PPP
```

```python
predicted_PPP = df[PPP]
predicted_PPP.head(20)
```

```python
# predicted_PPP = df[PPP]
confusion_matrix = [[0,0], 
                    [0,0]]

for actual, predicted in zip(actual_values,predicted_PPP):
    if actual == 1:
        if predicted == 1:
            confusion_matrix[0][0] += 1 #True Positive
        else:
            confusion_matrix[0][1] += 1 # False Negative
    else:
        if predicted == 1:
            confusion_matrix[1][0] += 1 #False Positive
        else:
            confusion_matrix[1][1] += 1 #True Negative
confusion_matrix
```

```python
disp = ConfusionMatrixDisplay(np.array(confusion_matrix), display_labels=['Cancer', 'Non Cancer'])
disp.plot()
```

```python
PIK = table1.head(2).index[1]
predicted_PIK = df[PIK]
predicted_PIK.head(20)
```

```python
confusion_matrix2 = [[0,0], 
                    [0,0]]

for actual, predicted in zip(actual_values, predicted_PIK):
    if actual == 1:
        if predicted == 1:
            confusion_matrix2[0][0] +=1 #TP
        else:
            confusion_matrix2[0][1] +=1 #FN
    else:
        if predicted == 1:
            confusion_matrix2[1][0] += 1 #False Positive
        else:
            confusion_matrix2[1][1] += 1 #True Negative
confusion_matrix2
```

```python
disp = ConfusionMatrixDisplay(np.array(confusion_matrix2), display_labels=['Cancer', 'Non Cancer'])
disp.plot()
```

## To determine which of the mutations is more useful to us, i decided to use the calculation of accuracy of true values to total values.

```python
def accuracy(TP, FN, FP, TN):
    return (TP+TN)/(TP+FN+FP+TN)

PPP_accuracy = accuracy(TP1, FN1, FP1, TN1)
PIK_accuracy = accuracy(TP2, FN2, FP2, TN2)

if PPP_accuracy > PIK_accuracy:
    print(f"The most accurate feature of the two is PPP_accuracy")
else:
    print(f"The most accurate feature of the two is PIK_accuracy with {PIK_accuracy}, while PPP has a lower accuracy of {PPP_accuracy}")
```

```python
df = df.set_index('class')
```

```python
diff_dict={}

for column in df.columns:
    TP = len(df[(df[column] == 1) & (df.index.str.startswith('C'))])
    FP = len(df[(df[column] == 1) & (df.index.str.startswith('NC'))])
    
    diff_dict[column] = TP-FP

max_mutation = max(diff_dict, key=diff_dict.get)
print([max_mutation, diff_dict[max_mutation]])
```

```python
Canc = df[df.index.str.startswith('C')].shape[0]
NCanc = df[df.index.str.startswith('NC')].shape[0]
total_samples = Canc + NCanc

percent_TP_values = []
percent_FP_values = []

for column in df.columns:
    TPi = df[df.index.str.startswith('C')][column].sum()
    FPi = df[df.index.str.startswith('NC')][column].sum()

    percent_TP_values.append(TPi / total_samples)
    percent_FP_values.append(FPi / total_samples)


difference = np.array(percent_TP_values) - np.array(percent_FP_values)

max_index = np.argmax(difference)
max_mutation = df.columns[max_index]

print(f"The mutation with the maximum value of %TP - %FP is: {max_mutation} with {difference[max_index]}")


```

```python
print(total_samples)
print(f"The mutation with the maximum value of %TP - %FP is: {max_mutation} with {difference[max_index]*100}")
```
