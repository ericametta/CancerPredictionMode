### shape attribute returns a tuple representing the dimensionality of a data frame. first element being row and second being column

```python
mutations_rows = mutations.shape[0]
mutations_cols = mutations.shape[1]
print(f"The data file contains {mutations_rows} patients and {mutations_cols} unique mutations")
```

### Lets select patient C1 and NC1 using iLoc

```python
mutations.iloc[1:3]
```

### next, lets filter the mutations data frame by class C1, then we drop the class column, leaving us with the mutations attribute. then we sum all the values on the second axis (1), since theyre either 1 or 0. the use of iloc to get scalar value of the sum from the pandas series

```python
mutations_C1 = mutations[mutations['class'] == 'C1'].drop(columns='class').sum(axis=1).iloc[0]
print(f"There are {mutations_C1} mutations under individual C1")
```

### repeat to get number of mutations in patient NC1

```python
mutations_NC1 = mutations[mutations['class'] == 'NC1'].drop(columns="class").sum(axis=1).iloc[0]
print(f"There are {mutations_NC1} mutations under individual NC1")
```

### to get the average number of mutations per individual, we can use the mean method

```python
average_mutations = mutations.drop(columns='class').sum(axis=1).mean()
print(average_mutations)
```

### to get the minimum and max number of mutations per individual, we can use the min and max function after computing the sum of the mutations across each column

```python
min_mutations = mutations.drop(columns='class').sum(axis=1).min()
print(min_mutations)
max_mutations = mutations.drop(columns='class').sum(axis=1).max()
print(max_mutations)
```

### the average number of individuals per mutation

```python
total_individual = mutations.drop(columns='class').sum()
print(total_individual) 
sum_total = mutations.drop(columns='class').sum().sum()
print(f"the sum of all individuals with mutations {sum_total}")
average_individual = mutations.drop(columns='class').sum().mean()
print(f"The average number of individuals per mutation is {average_individual}")
```

### the minimum and maximum number of individuals per mutation can be calculated by summing along the rows and using the min and max functions

```python
individuals_min = mutations.drop(columns='class').sum().min()
print(individuals_min)
individuals_max = mutations.drop(columns='class').sum().max()
print(individuals_max)
```

### To find the number of individuals with mutations in the KRAS and BRAF gene, we have to apply the 'filter' method to selct columns and use the like parameter to select the KRAS and BRAF columns.

```python
BRAF = mutations.filter(like='BRAF').sum().sum()
print(f"There are {BRAF} individuals who have mutations in the BRAF gene")

KRAS = mutations.filter(like='KRAS').sum().sum()
print(f"There are {KRAS} individuals who have mutations in the KRAS gene")
```
